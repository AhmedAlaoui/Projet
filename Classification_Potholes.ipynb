```
df1 = df.dropna(subset=['Etat_PCI'])
df1.count()
```

```
# add the category 3 for the empty 133 data 3 = zone interdite excepté pour livraison locale
df1.fillna(value = 3, axis=1, inplace=True)
```

```
# some encoding for all the categories
df1 = pd.concat([df1,pd.get_dummies(df1['affectation'], prefix='affectation')],axis=1)
df1 = pd.concat([df1,pd.get_dummies(df1['TYP_VOIE'], prefix='TYP_VOIE')],axis=1)
df1 = pd.concat([df1,pd.get_dummies(df1['CAMION'], prefix='CAMION')],axis=1)
df1 = pd.concat([df1,pd.get_dummies(df1['CLASSE'], prefix='CLASSE')],axis=1)
df1 = pd.concat([df1,pd.get_dummies(df1['SENS_CIR'], prefix='SENS_CIR')],axis=1)
df1.shape
```

```
#drop the table not needed
df1 = df1.drop(['NOM_VOIE','affectation','TYP_VOIE','CAMION','calls_2019',	'potholes_2019',	'freeze_thaw_2019',	'calls_2020'	,'potholes_2020',	'freeze_thaw_2020','snow_removal_2020', 'snow_removal_2019','length','CLASSE','SENS_CIR'], axis=1)
```

```
#df1 = df1.drop(['calls_2019',	'potholes_2019',	'freeze_thaw_2019',	'calls_2020'	,'potholes_2020',	'freeze_thaw_2020','snow_removal_2020', 'snow_removal_2019'], axis=1)
```

```
# we keep only the data with calls and potholes repareces
df_calls = df1[df1['calls_2018'] + df1['calls_2017'] +df1['calls_2016']>0]
df_potholes = df1[df1['potholes_2018'] + df1['potholes_2017'] +df1['potholes_2016']>0]
```

```
# append the date
df2 = df_calls.append(df_potholes, ignore_index=True)
#df3 = df_calls.append(df_potholes, ignore_index=True,verify_integrity=True)
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import RandomizedSearchCV
```

```
x = df2.drop(['Etat_PCI'], axis=1)
y = df2['Etat_PCI']
sc = StandardScaler()
x = sc.fit_transform(x)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.1, random_state=42)
```

```
for data in [y_train, y_test]:
    print(data.describe())
```

K Nearest Neighbors Classifier

```
n3 = KNeighborsClassifier(n_neighbors = 3)
n3.fit(x_train, y_train)
pred_n3 = n3.predict(x_test)
print(classification_report(y_test, pred_n3))
```

```
n5 = KNeighborsClassifier(n_neighbors = 5)
n5.fit(x_train, y_train)
pred_n5 = n5.predict(x_test)
print(classification_report(y_test, pred_n5))
```

Random Forest Classifier

```
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
pred_rf = rf.predict(x_test)
print(classification_report(y_test, pred_rf))
```

Decision Tree Classifier

```
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
pred_dt = dt.predict(x_test)
print(classification_report(y_test, pred_dt))
```

Stochastic Gradient Descent

```
sgd = SGDClassifier()
sgd.fit(x_train, y_train)
pred_sgd = sgd.predict(x_test)
print(classification_report(y_test, pred_sgd))
```

Trying to improve results

```
# number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start=50, stop=1000, num=10)]
# number of features to consider at every split
max_features = ['auto', 'sqrt']
# max number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num=10)] + [None]
# min number of samples required to split a node
min_samples_split = [2, 5, 10]
# min number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# method of selecting samples for training each tree
bootstrap = [True, False]

# random grid
random_grid = {'n_estimators': n_estimators,
              'max_features': max_features,
              'max_depth': max_depth,
              'min_samples_split': min_samples_split,
              'min_samples_leaf': min_samples_leaf,
              'bootstrap': bootstrap}

rf_optimized = RandomizedSearchCV(estimator=rf, param_distributions=random_grid, n_iter=100, cv=3, verbose=1, random_state=42)

# print(rf_optimized.best_params_)

rf_optimized.fit(x_train, y_train)
pred_optimized = rf_optimized.predict(x_test)
print(classification_report(y_test, pred_optimized))
```
